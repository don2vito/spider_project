```python
import pandas as pd
import requests
import json
import demjson
from jsonpath import jsonpath
```

```python
def get_info_month(isOnline,year,month):
    url = 'https://wechatapi.monitor.com.cn/api/services/app/MarketRead/MarketReadBulletin'
    data = {'category':'',
            'categories':[],
            'isOnline':isOnline,
            'year':year,
            'month':month,
            'isWeekYear':False}
    headers = {
    #            'Host': 'wechatapi.monitor.com.cn',
    #            'Connection': 'keep-alive',
    #            'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.143 Safari/537.36 MicroMessenger/7.0.9.501 NetType/WIFI MiniProgramEnv/Windows WindowsWechat',
               'content-type': 'application/json;charset=utf-8',
    #            'Referer': 'https://servicewechat.com/wxca2c15072ee86758/26/page-frame.html'
               }
    res = requests.post(url,data=json.dumps(data),headers=headers)
    json_data = demjson.decode(res.text)
    classify_list = jsonpath(json_data, '$..classify')
    groupName_list = jsonpath(json_data, '$..groupName')
    category_list = jsonpath(json_data, '$..category')
    val_list = jsonpath(json_data, '$..val')
    vol_list = jsonpath(json_data, '$..vol')
    tqVal_list = jsonpath(json_data, '$..tqVal')
    tqVol_list = jsonpath(json_data, '$..tqVol')
    sqTVal = jsonpath(json_data, '$..sqVal')
    sqTVol = jsonpath(json_data, '$..sqVol')
    valTb = jsonpath(json_data, '$..valTb')
    valHb = jsonpath(json_data, '$..valHb')
    volTb = jsonpath(json_data, '$..volTb')
    volHb = jsonpath(json_data, '$..volHb')
    list_content = [groupName_list,category_list,val_list,vol_list,tqVal_list,tqVol_list,sqTVal,sqTVol,valTb,valHb,volTb,volHb]
    df_jsonpath = pd.DataFrame(list_content)
    df_jsonpath = df_jsonpath.T
    df_jsonpath.rename(columns={
                                0:'中分类',
                                1:'细分类',
                                2:'本期销售额(万)',
                                3:'本期销售量(万)',
                                4:'同期销售额(万)',
                                5:'同期销售量(万)',
                                6:'上期销售额(万)',
                                7:'上期销售量(万)',
                                8:'零售额同比',
                                9:'零售额环比',
                                10:'零售量同比',
                                11:'零售量环比',
                                },inplace=True)
    df_jsonpath['零售量同比'] = df_jsonpath['零售量同比'].apply(lambda x:str(round(float(x) * 100,2)) + '%')
    df_jsonpath['零售量环比'] = df_jsonpath['零售量环比'].apply(lambda x:str(round(float(x) * 100,2)) + '%')
    df_jsonpath['零售额同比'] = df_jsonpath['零售额同比'].apply(lambda x:str(round(float(x) * 100,2)) + '%')
    df_jsonpath['零售额环比'] = df_jsonpath['零售额环比'].apply(lambda x:str(round(float(x) * 100,2)) + '%')
    df_jsonpath['本期销售额(万)'] = df_jsonpath['本期销售额(万)'].apply(lambda x:round(x / 10000,2))
    df_jsonpath['本期销售量(万)'] = df_jsonpath['本期销售量(万)'].apply(lambda x:round(x / 10000,2))
    df_jsonpath['同期销售额(万)'] = df_jsonpath['同期销售额(万)'].apply(lambda x:round(x / 10000,2))
    df_jsonpath['同期销售量(万)'] = df_jsonpath['同期销售量(万)'].apply(lambda x:round(x / 10000,2))
    df_jsonpath['上期销售额(万)'] = df_jsonpath['上期销售额(万)'].apply(lambda x:round(x / 10000,2))
    df_jsonpath['上期销售量(万)'] = df_jsonpath['上期销售量(万)'].apply(lambda x:round(x / 10000,2))
    return df_jsonpath.sort_values(by=['中分类','细分类'],ascending=True)
```

```python
isOnline = True # 是否线上
year = '2021' # 年
month ='2' # 月份
week = '' # 周次

if week == '':
    df = get_info_month(isOnline,year,month)
else:
    df = get_info_week(isOnline,year,week)

df
```

```python
# 持久化
if isOnline == True:
    if week == '':
        file_name = './' + year + '年' + month + '月' + '全国家电线上数据.xlsx'
        df.to_excel(file_name,sheet_name='全国家电线上数据',index=False)
        print(f'{file_name} 保存成功！！')
    else:
        file_name = './' + year + '年' + week + '周' + '全国家电线上数据.xlsx'
        df.to_excel(file_name,sheet_name='全国家电线上数据',index=False)
        print(f'{file_name} 保存成功！！')
else :
    if week == '':
        file_name = './' + year + '年' + month + '月' + '全国家电线下数据.xlsx'
        df.to_excel(file_name,sheet_name='全国家电线下数据',index=False)
        print(f'{file_name} 保存成功！！')
    else:
        file_name = './' + year + '年' + week + '周' + '全国家电线下数据.xlsx'
        df.to_excel(file_name,sheet_name='全国家电线下数据',index=False)
        print(f'{file_name} 保存成功！！')
```
